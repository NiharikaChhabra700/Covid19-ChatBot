```python
import pandas as pd
import numpy as np

df = pd.read_csv("output1.csv" ,header =None)
df = df.dropna()
df.head()
```

```python
df.columns=["text", "questions"]
```

```python
df
```

```python
df = df.drop(index=0)
df
```

```python
list = []
s=[]
xyz=[]
for ind in df.index: 
    list=df['questions'][ind]
    s=list.split(',')
    k=len(s)
    s[0]=s[0][1:]
    s[-1]=s[-1][:k-1]
    
    n=len(s)
    for i in range (n):
        s[i]= s[i].replace("'", "")
        xyz.append(s[i])


while(" " in xyz) :
    xyz.remove(" ")
print(xyz)
type(xyz[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances
```

```python
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(xyz).todense()
```

```python
print( vectorizer.vocabulary_ )
```

```python
for f in features:
    print( euclidean_distances(features[0], f) )
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(analyzer="char")
string = 'Is delta variant the fastest covid variant?'

sparse_matrix = tfidf_vectorizer.fit_transform([string]+xyz)
euclidean= euclidean_distances(sparse_matrix[0,:],sparse_matrix[1:,:])
pd.DataFrame({'euclidean':euclidean[0],'strings':xyz}).sort_values('euclidean',ascending=True)
```
