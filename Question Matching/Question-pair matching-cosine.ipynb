```python
import pandas as pd
import numpy as np

df = pd.read_csv("final_questions.csv" ,header =None)
df = df.dropna()
df.head()
```

```python
df.columns=["text", "questions"]
```

```python
df
```

```python
df = df.drop(index=0)
df
```

```python
list = []
s=[]
questions=[]
for ind in df.index: 
    list=df['questions'][ind]
    s=list.split(',')
    k=len(s)
    s[0]=s[0][1:]
    s[-1]=s[-1][:k-1]
    
    n=len(s)
    for i in range (n):
        s[i]= s[i].replace("'", "")
        questions.append(s[i])


while(" " in questions) :
    questions.remove(" ")
print(questions)
type(questions[0])
        
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np

string = 'Is delta variant the fastest covid variant?'

tfidf_vectorizer = TfidfVectorizer(analyzer="char")

sparse_matrix = tfidf_vectorizer.fit_transform([string]+questions)
cosine = cosine_similarity(sparse_matrix[0,:],sparse_matrix[1:,:])
```

```python
pd.DataFrame({'cosine':cosine[0],'strings':questions}).sort_values('cosine',ascending=False)
```
