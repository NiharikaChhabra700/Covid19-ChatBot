```python
import pandas as pd
import numpy as np

df = pd.read_table("Final_Annotation.txt" , sep='\t' ,header =None)

df = df.reindex(np.random.permutation(df.index))

df.reset_index(inplace=True, drop=True)

df.head()
```

```python
df.columns=["label", "tweet_message"]
```

```python
df.drop(df.loc[df["label"]==0].index, inplace=True)
```

```python
df
```

```python
df.dropna(subset = ["label"], inplace=True)
```

```python
df
```

```python
df['label'] = df['label'].astype(str)
df['tweet_message'] = df['tweet_message'].astype(str)
```

```python
df
```

```python
df['label']=df['label'].replace('1.0', 1)
```

```python
import csv
from pprint import pprint
from Questgen import main
qe= main.BoolQGen()
ques_list= []
# field_names = ['Questions', 'Count', 'Text']
for ind in df.index:
    payload = {
        "input_text": df['tweet_message'][ind]
    }
    output = qe.predict_boolq(payload)
    ques_list.append(output)
    
print(ques_list)    
# with open('out.csv', 'w') as csvfile:
#     writer = csv.writer(csvfile)
#     for obj in ques_list:
# #         writer.writerow(obj)
#         writer.writerows(obj)
        
keys = ques_list[0].keys()
print(keys)

a_file = open("output1.csv", "w", encoding='utf-8')
dict_writer = csv.DictWriter(a_file, keys)
dict_writer.writeheader()
dict_writer.writerows(ques_list)
a_file.close()
```
